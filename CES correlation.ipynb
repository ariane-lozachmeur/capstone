```python
import pandas as pd
import numpy as np
from math import *
import matplotlib.pyplot as plt
import matplotlib
import matplotlib
import scipy.stats
```

```python
def load_data(expression_file, mutation_file):
    expr = pd.read_table(expression_file)
    mut = pd.read_table(mutation_file)
    ces = expr.sum()[2:]
    data = mut[['Sample ID','CNA','Mutation Count']].merge(pd.DataFrame(ces,columns=['ces']),
                                                           how='inner',left_on='Sample ID',right_index=True)
    return data

def show_plots(data):
    plt.scatter(data['ces'], data['Mutation Count'])
    plt.ylabel('Mutation Count')
    plt.xlabel('CES')
    plt.title('Mutation and CES corralation')
    plt.show()
    print('CES and Mutation Spearman pvalue: ')
    data_nona = data[['ces','Mutation Count']].dropna()
    print(scipy.stats.spearmanr(data_nona['ces'],data_nona['Mutation Count']))

    plt.scatter(data['ces'], data['CNA'])
    plt.ylabel('CNA')
    plt.xlabel('CES')
    plt.title('CNA and CES corralation')
    plt.show()
    print('CES and CNA Sperrman pvalue: ')
    data_nona = data[['ces','CNA']].dropna()
    print(scipy.stats.spearmanr(data_nona['ces'],data_nona['CNA']))
```

# Adrenocortical carcinoma

```python
data = load_data('data/adrenocortical/ces_expr.txt','data/adrenocortical/mutation_cna.txt')
show_plots(data)
```

```python
data = load_data('data/bladder_carcinoma/cBioPortal_data.txt','data/bladder_carcinoma/Mutation Count vs. CNA.txt')
show_plots(data)
```
