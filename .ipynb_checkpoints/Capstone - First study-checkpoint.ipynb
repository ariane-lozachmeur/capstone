```python
data = pd.read_table('cBioPortal_data.txt')
data.head()
```

```python
def CES(patient,dataset,gene_list):
    sum = 0
    for g in gene_list:
        sum+= float(dataset[dataset['GENE_ID']==g][patient])
        
    return sum
```

```python
patients = []
for p in data.columns[2:]:
    ces = CES(p,data,data['GENE_ID'])
    patients.append([p,ces])

patients = pd.DataFrame(patients)
patients.columns = ['id','']
```
